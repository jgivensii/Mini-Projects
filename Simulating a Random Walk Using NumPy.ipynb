```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_steps = 1000
x_steps = np.random.randn(n_steps)
y_steps = np.random.randn(n_steps)
```

```python
y_walk = np.cumsum(y_steps)
x_walk = np.cumsum(x_steps) 
```

```python
walk_2d = np.column_stack((x_walk, y_walk))
```

```python
plt.plot(x_walk)
plt.show()
```

```python
plt.plot(y_walk)
plt.show()
```

```python
plt.plot(walk_2d[:,0],walk_2d[:,1])
plt.show()
print(f"The initial position is: ({walk_2d[0,0]},{walk_2d[0,1]})")
print(f"The final position is: ({walk_2d[-1,0]},{walk_2d[-1,1]})")
```
