```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sts
import seaborn as sns
```

```python
n_steps = 1000
x_steps = np.random.randn(n_steps)
y_steps = np.random.randn(n_steps)
```

```python
y_walk = np.cumsum(y_steps)
x_walk = np.cumsum(x_steps) 
```

```python
walk_2d = np.column_stack((x_walk, y_walk))
```

```python
plt.plot(x_walk)
plt.show()
```

```python
plt.hist(x_walk)
```

```python
plt.plot(y_walk)
plt.show()
```

```python
plt.hist(y_walk)
```

```python
plt.plot(walk_2d[:,0],walk_2d[:,1])
plt.show()
print(f"The initial position is: ({walk_2d[0,0]},{walk_2d[0,1]})")
print(f"The final position is: ({walk_2d[-1,0]},{walk_2d[-1,1]})")
```

```python
#Random Walk analysis
Ave_x_pos = np.average(x_walk)
Ave_y_pos = np.average(y_walk)
mode_x = sts.mode(x_walk).mode.item()
mode_y = sts.mode(y_walk).mode.item()
print(f"The average position for the random walk is: ({Ave_x_pos},{Ave_y_pos})")
print(f"The mode for the x coordinate: {mode_x} & mode for the y coordinate: {mode_y}")
#The average should be the center point on the plot. 
```
